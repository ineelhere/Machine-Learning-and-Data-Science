<h2>Intro to pyplot</h2>
<p><a class="reference internal" title="matplotlib.pyplot" href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot"><code class="xref py py-mod docutils literal notranslate"><span class="pre">matplotlib.pyplot</span></code></a>&nbsp;is a collection of functions that make matplotlib work like MATLAB. Each&nbsp;<code class="docutils literal notranslate"><span class="pre">pyplot</span></code>&nbsp;function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.</p>
<p>In&nbsp;<a class="reference internal" title="matplotlib.pyplot" href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot"><code class="xref py py-mod docutils literal notranslate"><span class="pre">matplotlib.pyplot</span></code></a>&nbsp;various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the&nbsp;<em>axes</em>&nbsp;<a class="reference internal" href="https://matplotlib.org/3.3.1/tutorials/introductory/usage.html#figure-parts"><span class="std std-ref">part of a figure</span></a>&nbsp;and not the strict mathematical term for more than one axis).</p>

## A matplotlib workflow

```python
# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(15,5))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show
fig.savefig("example.png")
```

**Bar Plot**

```python
#let us create a dictionary of our own - it has mountains and their heights
mountains = { 'Mount Everest ' : 8848,
                'K2 ' : 8611,
                'Kangchenjunga ' : 8586,
                'Lhotse ' : 8516,
                'Makalu ' : 8485,
                'Cho Oyu ' : 8201,
                'Dhaulagiri ' : 8167,
                'Manaslu ' : 8163,
                'Nanga Parbat ' : 8126,
                'Annapurna ' : 8091,
                'Gasherbrum I(Hidden peak; K5) ' : 8080,
                'Broad Peak ' : 8051,
                'Gasherbrum II(K4) ' : 8035,
                'Shishapangma ' : 8027 }
mountains
```

```python
fig, ax = plt.subplots()
ax.set(title="World's Heighest Mountains", ylabel="Height(ft)", xlabel="Names")
ax.bar(mountains.keys(), mountains.values());
```

```python
#see that the X axis labels are merging? Let's fix that
fig, ax = plt.subplots()
ax.set(title="World's Heighest Mountains", ylabel="Height(ft)", xlabel="Names")
plt.xticks(rotation=90) #this is how we change the orientation of the label
ax.bar(mountains.keys(), mountains.values());
```

```python
#Isn't the above plot a bit awkward? Here is a fix -
fig, ax = plt.subplots()
ax.set(title="World's Heighest Mountains", ylabel="Height(ft)", xlabel="Names")
plt.xticks(rotation=90) #this is how we change the orientation of the label
ax.barh(list(mountains.keys()), list(mountains.values())); #
```

barh() function in axes module of matplotlib library is used to make a horizontal bar plot.

# Subplot Options
**Option 1**

```python
#set up multiple plots
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,5))
```

```python
#set up multiple plots 
fig, ((ax1,ax2), (ax3,ax4), (ax5,ax6)) = plt.subplots(nrows=3,
                                          ncols=2,
                                          figsize=(10,5))
```

```python
#so let's have fun with them B-)
#set up multiple plots
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(10,10))
#Plot to different axes

ax1.barh(list(mountains.keys()), list(mountains.values()));
ax2.plot([2,4,3,1,6,7], [8,9,10,11,12,16]);
ax3.scatter(data, np.sin(data));
ax4.hist(data);
```

**Option 2**

```python
#set up the plot
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10,10))

# Now plot to each different INDEX
# ax[row, column]

ax[0,0].barh(list(mountains.keys()), list(mountains.values()));
ax[0,1].plot([2,4,3,1,6,7], [8,9,10,11,12,16]);
ax[1,0].scatter(data, np.sin(data));
ax[1,1 ].hist(data);
```
