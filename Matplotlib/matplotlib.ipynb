<h2>Intro to pyplot</h2>
<p><a class="reference internal" title="matplotlib.pyplot" href="https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot"><code class="xref py py-mod docutils literal notranslate"><span class="pre">matplotlib.pyplot</span></code></a>&nbsp;is a collection of functions that make matplotlib work like MATLAB. Each&nbsp;<code class="docutils literal notranslate"><span class="pre">pyplot</span></code>&nbsp;function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.</p>

# Important
* Using pyplot method is OK for simple plots.
* For more complicated plots,it is better to use the Object Oriented approach.


<div id="a-note-on-the-object-oriented-api-vs-pyplot" class="section">
<h2>A note on the Object-Oriented API vs. Pyplot</h2>
<p>Matplotlib has two interfaces. The first is an object-oriented (OO) interface. In this case, we utilize an instance of&nbsp;<a class="reference internal" title="matplotlib.axes.Axes" href="https://matplotlib.org/3.2.1/api/axes_api.html#matplotlib.axes.Axes"><code class="xref py py-class docutils literal notranslate"><span class="pre">axes.Axes</span></code></a>&nbsp;in order to render visualizations on an instance of&nbsp;<a class="reference internal" title="matplotlib.figure.Figure" href="https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.figure.Figure.html#matplotlib.figure.Figure"><code class="xref py py-class docutils literal notranslate"><span class="pre">figure.Figure</span></code></a>.</p>
<p>The second is based on MATLAB and uses a state-based interface. This is encapsulated in the&nbsp;<a class="reference internal" title="matplotlib.pyplot" href="https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot"><code class="xref py py-mod docutils literal notranslate"><span class="pre">pyplot</span></code></a>&nbsp;module. See the&nbsp;<a class="reference internal" href="https://matplotlib.org/3.2.1/tutorials/introductory/pyplot.html"><span class="doc">pyplot tutorials</span></a>&nbsp;for a more in-depth look at the pyplot interface.</p>
<p>Most of the terms are straightforward but the main thing to remember is that:</p>
<ul class="simple">
<li>The Figure is the final image that may contain 1 or more Axes.</li>
<li>The Axes represent an individual plot (don't confuse this with the word "axis", which refers to the x/y axis of a plot).</li>
</ul>
<p>We call methods that do the plotting directly from the Axes, which gives us much more flexibility and power in customizing our plot.</p>
<div class="admonition note">
<p class="first admonition-title">Note</p>
<p class="last">In general, try to use the object-oriented interface over the pyplot interface.</p>
</div>

## So, the Object Oriented approach for plots -

```python
#get the entries where age>50
up50 = hd[hd.age>50]
up50
```

```python
#now a scatter plot -- Pyplot method

up50.plot(kind='scatter', 
             x='age', 
             y='chol', 
             c='target',
             figsize = (15,5));
```

```python
#same as above, but OO method

fig, axis = plt.subplots(figsize=(15,5))
up50.plot(kind='scatter', 
             x="age", 
             y="chol", 
             c='target', 
             ax=axis);
# ax.set_xlim([30, 100]); #Set the x-axis view limits.
```

```python
#Another one!

#first things first
fig, (axis1, axis2) = plt.subplots(nrows = 2,
                                   ncols = 1,
                                   figsize = (15,15))

#FOR 1ST SUBPLOT
#plot the data
figure = axis1.scatter(x = up50.age,
                     y = up50.chol,
                     c = up50.target);

#make the plot more readable
axis1.set(title = "Heart Disease - A plot to show the relationship b/w age, cholesterol levels and the targets",
          xlabel = "age",
          ylabel = "cholesterol");

#add a legend/index/reference (I like to call it legend)
axis1.legend(*figure.legend_elements(), title = "target");

#add a horizontal line across the axes
axis1.axhline(up50.chol.mean(),
            linestyle = "--"); #The line signifies the mean cholesterom level

#FOR 2ND SUBPLOT
#plot the data
figure = axis2.scatter(x = up50.thalach,
                     y = up50.trestbps,
                     c = up50.target);

#make the plot more readable
axis2.set(title = "Heart Disease - A plot to show the relationship b/w maximum heart rate achieved, resting blood pressure and the targets",
          xlabel = "maximum heart rate achieved",
          ylabel = "resting blood pressure"); #thalach means maximum heart rate achieved and trestbps mean resting blood pressure

#add a legend/index/reference (I like to call it legend)
axis2.legend(*figure.legend_elements(), title = "target");

#add a horizontal line across the axes
axis2.axhline(up50.trestbps.mean(),
            linestyle = "--"); #The line signifies the mean resting blood pressure

#now add a grand title (suptitle)
fig.suptitle("A graphical representation and analysis of the heart diseases dataset \n for patients aged above 50 years", 
               fontsize = 20, fontweight = "bold");
```

## Change colors using 'cmaps'
* Documentation - https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

```python
#set the style
plt.style.use("fast")

#first things first
fig, (axis1, axis2) = plt.subplots(nrows = 2,
                                   ncols = 1,
                                   figsize = (12,12))

#FOR 1ST SUBPLOT
#plot the data
figure = axis1.scatter(x = up50.age,
                     y = up50.chol,
                     c = up50.target,
                     cmap = "winter" ); #cmap goes here

#make the plot more readable
axis1.set(title = "Heart Disease - A plot to show the relationship b/w age, cholesterol levels and the targets",
          xlabel = "age",
          ylabel = "cholesterol");

#add a legend/index/reference (I like to call it legend)
axis1.legend(*figure.legend_elements(), title = "target");

#add a horizontal line across the axes
axis1.axhline(up50.chol.mean(),
            linestyle = "--"); #The line signifies the mean cholesterom level

#FOR 2ND SUBPLOT
#plot the data
figure = axis2.scatter(x = up50.thalach,
                     y = up50.trestbps,
                     c = up50.target,
                       cmap="summer"
                      ); #cmap goes here

#make the plot more readable
axis2.set(title = "Heart Disease - A plot to show the relationship b/w maximum heart rate achieved, resting blood pressure and the targets",
          xlabel = "maximum heart rate achieved",
          ylabel = "resting blood pressure"); #thalach means maximum heart rate achieved and trestbps mean resting blood pressure

#add a legend/index/reference (I like to call it legend)
axis2.legend(*figure.legend_elements(), title = "target");

#add a horizontal line across the axes
axis2.axhline(up50.trestbps.mean(),
            linestyle = "--"); #The line signifies the mean resting blood pressure

#now add a grand title (suptitle)
fig.suptitle("A graphical representation and analysis of the heart diseases dataset \n for patients aged above 50 years \n", 
               fontsize = 20, fontweight = "bold");
```
